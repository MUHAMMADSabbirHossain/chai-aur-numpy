## Advance operation with Business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [retaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], # Paradise Biryani
    [2, 120000, 140000, 160000, 190000], # Beijing Bites
    [3, 200000, 230000, 260000, 300000], # Pizza Hub
    [4, 180000, 210000, 240000, 270000], # Burger Point
    [5, 160000, 185000, 205000, 230000] # Chai Point
])

print("==== Zomato sales analysys ====")
print("\nSales data shape: ", sales_data.shape)
print("\nSample data for 1st 3 restautants: \n", sales_data[:3])
print("\nSample data for 1st 3 restautants: \n", sales_data[:, 1:]) # without restautant id
```

```python
# toatl sales per year
print(np.sum(sales_data, axis = 0))
year_total = np.sum(sales_data[:, 1:], axis = 0)
print("Yearly total sales: ", year_total)

# Minimum sales per restaurant 
min_sales = np.min(sales_data[:, 1:], axis = 1)
print("Minimum sales per restaurant: ", min_sales)

# Maximum sales per year
max_sales = np.max(sales_data[:, 1:], axis = 0)
print("Maximum sales per year: ", max_sales)

# Average sales per year
avg_sales = np.mean(sales_data[:, 1:], axis = 1)
print("Average sales per year: ", avg_sales)
```

```python
cumsum = np.cumsum(sales_data[:, 1:], axis = 1)
print("Cumulative sum: \n", cumsum)

plt.figure(figsize = (10, 6))
plt.plot(np.mean(cumsum, axis = 0))
plt.title("Average cumulative sales accross all restaurants")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector addition: ", vector1 + vector2)
print('\nMultiplication vector: ', vector1 * vector2)
print("\nDot Product: ", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) /
                  (
                      np.linalg.norm(vector1) * np.linalg.norm(vector2))
                    )
print("Angle between vectors: ", angle)
```

```python
restaurant_types = np.array(['biryani', 'chinese', 'pizza', 'burger', 'cafe'])
vectorized_upeer = np.vectorize(str.upper)

print("Vectorized Upper: ", vectorized_upeer(restaurant_types))
```
